```python
#установка библиотеки и подключение всех необходимых библиотек
from nltk.tokenize import word_tokenize
from string import punctuation
from nltk.stem.snowball import RussianStemmer
import nltk
import pandas as pd
import numpy as np
import pymorphy2


nltk.download('punkt')
nltk.download('stopwords')
```

```python
#1.Придумать 2-3 предложения, содержащие знаки препинания, заглавные буквы.
string = 'Кошки обладают уникальной способностью оставаться спокойными в любой ситуации, что делает их идеальными спутниками в жизни. Кошки - настоящие чудеса природы!'
```

```python
#2.Провести предобработку на тексте из п.1. Вывести результат получившейся предобработки.
#токенизируем текст и записываем в отдельную переменную
tokens = word_tokenize(string)
print('Токенизированный текст:',tokens)

#удаляем пунктуацию
punctuations = list(punctuation)
tokens_without_punct = [i for i in tokens if i not in punctuations]
print('Текст без пунктуации:', tokens_without_punct)

#понижаем регистр слов
low_tokens = [i.lower() for i in tokens_without_punct]
print('Текст с понижением регистра:', low_tokens)

#удаляем стоп-слова
stopwords = nltk.corpus.stopwords.words('russian') #массив русских стоп-слов
words_without_stop = [i for i in low_tokens if i not in stopwords]
print('Текста без стоп слов:', words_without_stop)

#Лематизация текста
morph = pymorphy2.MorphAnalyzer() #переменная функции лемматизации
lemms = [morph.parse(i)[0].normal_form for i in words_without_stop]
print('Лематизированный текста', lemms)

#Стемминг текста
stemmer = RussianStemmer() #переменная функции стемминга
stems = [stemmer.stem(i) for i in lemms]
print('Выделенные сновы слов: ',stems)
```

```python
#3.Загрузить полученный файл из предыдущего домашнего задания.
db = pd.read_csv('/content/result.csv', sep = ',')
db.head()
```

```python
#4-5.Выполнить предобработку всех текстов из загрузочного файла и записать в отдельную колонку датафрейма
def preprocessing(row): #функция предобработки текста
  t_prep = word_tokenize(row)

  t_w_punct = [i for i in t_prep if i not in punctuations]
  low_t = [i.lower() for i in t_w_punct]
  w_w_stop = [i for i in low_t if i not in stopwords]

  lemms_prep = [morph.parse(i)[0].normal_form for i in w_w_stop]

  return lemms_prep

def join_strs(strs): #функция вывода массива значений в строку
    return ' '.join(strs)

def itog_completion_with_stemp_and_lemms(row): #функция заполнения столбца с использованием функции предобработки + стемпинг и вывода
  stroca = row['Transcription']
  mass = preprocessing(stroca)
  mass = [stemmer.stem(i) for i in mass]
  itog_stroca = join_strs(mass)

  return itog_stroca

def itog_completion_with_lemms(row): #функция заполнения столбца с использованием функции предобработки и вывода
  stroca = row['Transcription']
  mass = preprocessing(stroca)
  itog_stroca = join_strs(mass)

  return itog_stroca

db['New_Transcription'] = db.apply(itog_completion_with_stemp_and_lemms, axis=1)
db.head()
db['New_Transcription_only_lemms'] = db.apply(itog_completion_with_lemms, axis=1)
db.head()
```

```python
#6.Сохранить полученный датафрейм c новым названием result_laba2
db.to_csv('/content/result_laba2.csv', index=False)
```
