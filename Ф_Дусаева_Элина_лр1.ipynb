```python
#импортируем необходимые библиотеки
import pandas as pd
import numpy as np

#1.подгружаем файл первым способом, данные открыты с помощью DataFrame
db = pd.read_csv('/content/Manual_transcription clear with label.csv', sep = ';')
db.head()
```

```python
#2.Удалить колонку Name_audio
db = db.drop(columns = 'Name_audio')
db.head()
```

```python
#3.Переименовать колонку 'label' в 'Emotion'
db = db.rename(columns = {'label':'Emotion'})
db.head()
```

```python
#функция заполнения столбца по условию
#гнев, грусть, отвращение, страх - негативный сентимент;
#радость, удивление - позитивный сентимент, нейтраль -* нейтральное состояние*
def f(row):
 if (row['Emotion'] == 'гнев' or row['Emotion'] == 'грусть' or row['Emotion'] == 'отвращение' or row['Emotion'] == 'страх'):
   val = 'негативный сентимент'
 elif (row['Emotion'] == 'радость' or row['Emotion'] == 'удивление'):
   val = 'позитивный сентимент'
 elif (row['Emotion'] == 'нейтраль'):
   val = 'нейтральное состояние'
 else:
   val = 'неопознанно'
 return val

#4-5.Добавление колонки 'Sentiment' и заполнение при помощи функции
db['Sentiment'] = db.apply(f, axis=1)
db.head()
```

```python
#6.Проверка, все эмоции и состояния преобразовались корректно, случая 'неопознанно' не было
np.unique(db['Sentiment'])
```

```python
#7.Преобразовать эмоции в цифры (с помощью словаря). Записать в 'Emotion digits'
emo_digits = {'гнев': 0, 'грусть': 1, 'отвращение': 2, 'страх': 3,
              'радость': 4, 'удивление': 5, 'нейтраль': 6}

#Добавление колонки 'Emotion digits' и заполнение при помощи словаря, используя функцию map
db['Emotion digits'] = db['Emotion'].map(emo_digits)
db.head()
```

```python
#8.Преобразовать сентимент в цифры. (аналогично 7). Записать в отдельную колонку 'Sentiment digits'
sent_digits = {'негативный сентимент': 0, 'позитивный сентимент': 1, 'нейтральное состояние': 2}

db['Sentiment digits'] = db['Sentiment'].map(sent_digits)
db.head()
```

```python
#9.Сохранить измененный файл в csv с новым названием result
db.to_csv('/content/result.csv', index=False)
```
